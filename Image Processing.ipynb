## Geometic Transformation on images

### Resize Images

```python
img = cv2.imread('stop.png', 0)

# resize operation
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

print(img.shape)
print(res.shape)
```

### Translation

![alt tex](http://docs.opencv.org/3.0-beta/_images/math/43184a6b18ed02f11b98422bfb54ab3e48f8c08b.png)

```python
rows, cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (cols, rows))

f, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap='gray')
ax[1].imshow(dst, cmap='gray')
```

### Rotation

![alt tex](http://docs.opencv.org/3.0-beta/_images/math/bd671e57f9f5c4cd5fdd34fbd3244e7e6013354a.png)

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

f, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap='gray')
ax[1].imshow(dst, cmap='gray')
```
